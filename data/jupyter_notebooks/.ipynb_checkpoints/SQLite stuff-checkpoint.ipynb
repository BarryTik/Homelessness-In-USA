```python
def create_connection(db_file):
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        print(sqlite3.version)
    except Error as e:
        print(e)
    return conn
```

```python
def create_table(conn, create_table_sql):
    try:
        c = conn.cursor()
        c.execute(create_table_sql)
    except Error as e:
        print(e)
```

```python
def main():
    database = r"HomelessOwl_db.db"
    
    sql_create_hic_table = """CREATE TABLE IF NOT EXISTS hic (
                                        id integer PRIMARY KEY,
                                        state varchar(2),
                                        beds integer,
                                        year integer
                                    );"""
    sql_create_homeless_table = """CREATE TABLE IF NOT EXISTS homeless (
                                        id integer PRIMARY KEY,
                                        state varchar(2),
                                        overallhomeless integer,
                                        shelteredhomeless integer,
                                        unshelteredhomeless integer,
                                        individualhomeless integer,
                                        infamilyhomeless integer,
                                        chronichomeless integer,
                                        veteranhomeless integer,
                                        year integer
                                    );"""
    sql_create_population_table = """CREATE TABLE IF NOT EXISTS population (
                                        id integer PRIMARY KEY,
                                        state varchar(30),
                                        region integer,
                                        pop2007 integer,
                                        pop2008 integer,
                                        pop2009 integer,
                                        pop2010 integer,
                                        pop2011 integer,
                                        pop2012 integer,
                                        pop2013 integer,
                                        pop2014 integer,
                                        pop2015 integer,
                                        pop2016 integer,
                                        pop2017 integer,
                                        pop2018 integer,
                                        pop2019 integer
                                    );"""
    sql_create_pit_table = """CREATE TABLE IF NOT EXISTS pit (
                                    id integer PRIMARY KEY,
                                    state varchar(2),
                                    homelessness integer,
                                    year integer
                                );"""
    conn = create_connection(database)
    
    if conn is not None:
        create_table(conn, sql_create_hic_table)
        create_table(conn, sql_create_homeless_table)
        create_table(conn, sql_create_population_table)
        create_table(conn, sql_create_pit_table)
    else:
        print("Error! cannot create the database connection.")
```

```python
if __name__ == '__main__':
    main()
```

```python
def create_hic(conn, hic):
    sql = ''' INSERT INTO hic(id, state, beds, year)
                VALUES(?,?,?,?)'''
    cur = conn.cursor()
    cur.execute(sql, hic)
```

```python
def create_homeless(conn, homeless):
    sql = ''' INSERT INTO homeless(id, state, overallhomeless, shelteredhomeless, unshelteredhomeless, individualhomeless, infamilyhomeless, chronichomeless, veteranhomeless, year)
                VALUES(?,?,?,?,?,?,?,?,?,?)'''
    cur = conn.cursor()
    cur.execute(sql, homeless)
```

```python
def create_pit(conn, pit):
    sql = '''INSERT INTO pit(id, state, homelessness, year)
                VALUES(?,?,?,?)'''
    cur = conn.cursor()
    cur.execute(sql, pit)
```

```python
def create_population(conn, population):
    sql = '''INSERT INTO population(id, state, region, pop2007, pop2008, pop2009, pop2010, pop2011, pop2012, pop2013, pop2014, pop2015, pop2016, pop2017, pop2018, pop2019)
                VALUES(?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?)'''
    cur = conn.cursor()
    cur.execute(sql, population)
```

```python
import pandas as pd
```

```python
population_df = pd.read_csv("population_census.csv")
```

```python
population_df.head()
```

```python
database = r"HomelessOwl_db.db"
conn = create_connection(database)
with conn:
    for row in population_df.iterrows():
        data = (f"{row[1].ID}", f"{row[1].STATE}", f"{row[1].Regions}", f"{row[1].POPESTIMATE2007}", f"{row[1].POPESTIMATE2008}", f"{row[1].POPESTIMATE2009}", f"{row[1].POPESTIMATE2010}", f"{row[1].POPESTIMATE2011}", f"{row[1].POPESTIMATE2012}", f"{row[1].POPESTIMATE2013}", f"{row[1].POPESTIMATE2014}", f"{row[1].POPESTIMATE2015}", f"{row[1].POPESTIMATE2016}", f"{row[1].POPESTIMATE2017}", f"{row[1].POPESTIMATE2018}", f"{row[1].POPESTIMATE2019}")
        create_population(conn, data)
```
