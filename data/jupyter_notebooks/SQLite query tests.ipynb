```python
import sqlite3
from sqlite3 import Error
```

```python
def create_connection(db_file):
    conn = None
    try:
        conn = sqlite3.connect(db_file)
    except Error as e:
        print(e)
        
    return conn
```

```python
def select_all_pit(conn):
    cur = conn.cursor()
    cur.execute("SELECT * FROM pit")
    
    rows = cur.fetchall()
    
    for row in rows:
        print(row)
```

```python
database = r"HomelessOwl_db.db"
conn = create_connection(database)
with conn:
    select_all_pit(conn)
```
